```python
invalid_year=(~autos["registration_year"].between(1900,2016)).sum()/autos.shape[0]
print(invalid_year)
```

```python
autos["brand"].unique()
autos["brand"].value_counts().head(20)
over_5 = autos["brand"].value_counts()/autos.shape[0]
print(over_5)
```

The top 5 brand is German company and it's over 50% of the whole list. Many brands are lower than 5% and we will focus on the brand which has high proportion.

```python
brand_counts = autos["brand"].value_counts(normalize = True)
common_brands = brand_counts[brand_counts>0.05].index
print(common_brands)
```

```python
brand_price = {}
for row in common_brands:
    brand_only = autos[autos["brand"] == row]
    mean_price = brand_only["price"].mean()
    brand_price[row] = int(mean_price)
    
print(brand_price)
```

In the result, it shows that Mercedes is the most expensive and Opel is the most affordable brand.

```python
milage_counts = autos["obometer_km"].value_counts(normalize = True)
print(milage_counts)
```

```python
brand_milage = {}
for row in common_brands:
    brand_only = autos[autos["brand"]== row]
    mean_milage = brand_only["obometer_km"].mean()
    brand_milage[row]= float(mean_milage)
    
print(brand_milage)
```

```python
brand_price_series = pd.Series(brand_price).sort_values(ascending = False)
brand_milage_series = pd.Series(brand_milage).sort_values(ascending=False)
brand_price_dataframe = pd.DataFrame(brand_price_series, columns = ['mean_price'])
brand_milage_dataframe = pd.DataFrame(brand_milage_series, columns = ['mean_milage'])
print(brand_price_dataframe)
print(brand_milage_dataframe)
```
